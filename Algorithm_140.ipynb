```python
n = 1029
m = 1071

def solution(n, m):
    
    answer = []
    gcf = []
    for_reduce_iter = min(n,m)
    for i in range(1, for_reduce_iter+1):
        if (n % i ==0) & (m % i ==0):
            gcf.append(i)
    gcf = max(gcf)
    
    lcf = []
    lcf.append(n // gcf)
    lcf.append(m // gcf)
    lcf.append(gcf)
    def multyple_all(x):
        return eval('*'.join([ str(x) for x in lcf ]))
    
    answer.append(gcf)
    answer.append(multyple_all(lcf))
    return answer
solution(n, m)
```
