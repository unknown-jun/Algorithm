가장 적은 연산을 하려면 최대 공약수를 구할땐 두 숫자 중 작은 수로 iteration을 돌려야 함

```python
n = 78696
m = 19332
```

```python
gcf = []
for_reduce_iter = min(n,m)
for i in range(1, for_reduce_iter+1):
    if (m % i ==0) & (n%i==0):
        gcf.append(i)
max(gcf)
```

최소공배수는 두 수를 최대 공약수로 나눈 후, 나머지를 모두 곱하면 됨

```python
lcf = []

lcf.append(n // max(gcf))
lcf.append(m // max(gcf))
lcf.append(max(gcf))
lcf
```

리스트 안의 원소들을 곱하기 위해서 하나하나의 원소를 str로 변환 후 eval 함수를 적용

```python
'*'.join([ str(x) for x in lcf ])
```

```python
def multyple_all(x):
    return eval('*'.join([ str(x) for x in lcf ]))
multyple_all(lcf)
```

함수화

```python
n = 1029
m = 1071

def solution(n, m):
    
    answer = []
    gcf = []
    for_reduce_iter = min(n,m)
    for i in range(1, for_reduce_iter+1):
        if (n % i ==0) & (m % i ==0):
            gcf.append(i)
    gcf = max(gcf)
    
    lcf = []
    lcf.append(n // gcf)
    lcf.append(m // gcf)
    lcf.append(gcf)
    def multyple_all(x):
        return eval('*'.join([ str(x) for x in lcf ]))
    
    answer.append(gcf)
    answer.append(multyple_all(lcf))
    return answer
solution(n, m)
```
