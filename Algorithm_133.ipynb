```python
def solution(original_string):
    
    answer = ''
    split_string = []
    for i in original_string.split(" "):
        split_string.append(i + ' ')
    
    for strings in split_string:
        cnt = 0
        
        for alpha in strings:
            cnt+=1
            if cnt % 2 == 1:
                answer += alpha.upper()
            else:
                answer += alpha.lower()
    return answer[:-1]

solution(original_string)
```
