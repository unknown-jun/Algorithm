# 공백을 기준으로 나누기
- 맨 마지막에 공백을 추가하는 이유는 강제적으로 공백을 추가하여
- 마지막 리턴형식을 맞추기 위함

```python
original_string = "try hello world"

split_string = []
for i in original_string.split(" "):
    split_string.append(i + ' ')
split_string
```

# 리스트에 담긴 문자에 홀수번째 문자를 대문자로 변환

```python
cnt = 0

for strings in split_string:
    for alpha in strings:
        cnt+=1
        print(cnt)
```

```python
for strings in split_string:
    cnt = 0
    for alpha in strings:
        cnt+=1
        if cnt % 2 == 1:
            print(alpha.upper())
        else:
            print(alpha)
```

# answer에 하나씩 문자열로 더하기
- cnt를 초기화할때 공백을 넣어줘야 한다

```python
answer = ''
for strings in split_string:
    cnt = 0
    answer += ' '
    for alpha in strings:
        cnt+=1
        if cnt % 2 == 1:
            answer += alpha.upper()
        else:
            answer += alpha.lower()
            
print(answer)
```

# 함수식화

```python
def solution(original_string):
    
    answer = ''
    split_string = []
    for i in original_string.split(" "):
        split_string.append(i + ' ')
    
    for strings in split_string:
        cnt = 0
        
        for alpha in strings:
            cnt+=1
            if cnt % 2 == 1:
                answer += alpha.upper()
            else:
                answer += alpha.lower()
    return answer[:-1]

solution(original_string)
```
