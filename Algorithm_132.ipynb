```python
n = 12

for i in range(1,n+1):
    print(i)
```

약수는 나눴을 때 0이 되는 수

```python
tmp = []
for i in range(1, n+1):
    if n % i ==0:
        tmp.append(i)
sum(tmp)
```

list comprehension을 이용하여 한줄로 표현하기

```python
n = 12
sum([ i for i in range(1, n+1) if n % i ==0])
```

n의 반만 나누면 약수를 모든 수를 찾을 수 있음

```python
for i in range(1, (n//2)+1):
    if n % i ==0:
        print(i)
```

함수화

```python
def solution(n):

    return n+ sum([ i for i in range(1, (n//2)+1) if n % i ==0])

n = [12, 5]
answer = [28, 6]

for i,z in zip(n, answer):
    if solution(i)== z:
        print("정답")
```
