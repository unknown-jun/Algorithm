```python
def solution(arr, divisor):
    answer = []
    
    for i in arr:
        if i % divisor==0:
            answer.append(i)
            
    if len(answer)==0:
        answer.append(-1)
        
    return sorted(answer)
```

```python
arr=[[5, 9, 7, 10],[2, 36, 1, 3],[3,2,6]]
divisor = [5, 1, 10]
answer = [[5, 10],	[1, 2, 3, 36], 	[-1]]
for i,z,t in zip(arr,divisor, answer):
    if t == solution(i,z):
        print('')
    else:
        print('')
```
