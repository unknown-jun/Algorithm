3. 일주일 단위로의 함수는 생성이 되었다. 이젠 달 기준으로 바뀌는 함수를 생성해야 함

```python
a = 5
b = 24
def solution(a, b):
    
    anwer=''
    answer_m = 0
    day = [ 'THU','FRI','SAT','SUN','MON','TUE','WED' ]
    
    month = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    for i in range(a-1):
        if a == 1:
            continue
        answer_m += month[i]

    answer = day[(answer_m+b)%7]

    return answer

solution(a, b)
```

4. 좀 더 간단하게 할수 있지 않을까?? for 문을 제외해보자

```python
a = 5
b = 24
def solution(a, b):
    
    anwer=''
    day = [ 'THU','FRI','SAT','SUN','MON','TUE','WED' ]
    month = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    answer = day[(sum(month[:a-1])+b)%7]

    return answer

solution(a, b)
```
