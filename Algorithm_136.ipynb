# 한줄로 풀고 싶었음

```python
n=118372
def solution(n):
    return int(''.join(sorted(str(n), reverse=True)))
solution(n)
```

# 가독성 높이기 위해서 한 줄 함수 풀어헤치기

```python
n= 118372
print(n)

n= str(n)
print('sorted 함수를 쓰기 위해서 str로 형변환:', n)

n = sorted(n, reverse=True)
print('sorted 함수의 reverse 속성을 써서 내림차순으로 정렬\n',n)

n = ''.join(n)
print('list으로 묶인 하나하나의 요소를 하나로 풀어 str로 형변환까지 하기: ',n)

n = int(n)
print('int로 형변환하여 문제에서 원하는 형식으로의 제출', n)
```
