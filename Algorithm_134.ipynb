# 가장 단순한 방식
- for문을 이용하여 튜플안에 있는 자릿수를 뽑아내어 초기값에 더함

```python
def solution(n):
    
    n = str(n)
    n = tuple(n)

    sum_num = 0
    for num in n:
        sum_num += int(num)

    return sum_num
```

```python
N=[123, 987]
answer = [6, 24]

for i, z in zip(N, answer):
    if solution(i) == z:
        print("정답")
```

# 재귀함수를 쓰는 방식
- 10을 나누고 그 몫을 취한 다음에
- 재귀함수에 n 자기자신이 10보다 작을 때 종료 조건으로 취한다.

```python
123 // 10
```

```python
def solution(n):
    
    if n < 10:
        return n
    return n % 10 + solution(n//10)
```

```python
N=[123, 987]
answer = [6, 24]

for i, z in zip(N, answer):
    if solution(i) == z:
        print("정답")
```
