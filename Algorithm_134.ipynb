```python
def solution(n):
    
    n = str(n)
    n = tuple(n)

    sum_num = 0
    for num in n:
        sum_num += int(num)

    return sum_num
```

```python
N=[123, 987]
answer = [6, 24]

for i, z in zip(N, answer):
    if solution(i) == z:
        print("")
```

```python
def solution(n):
    
    if n < 10:
        return n
    return n % 10 + solution(n//10)
```

```python
N=[123, 987]
answer = [6, 24]

for i, z in zip(N, answer):
    if solution(i) == z:
        print("")
```
