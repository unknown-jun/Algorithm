```python
x = 132
def solution(x):
    str_x = str(x)
    sep_x = [ i for i in str_x ]
    sum_x = sum(map(int,sep_x))
    if x % sum_x == 0:
        return True
    else:
        return False
solution(x)
```

```python
x = [10, 11, 12, 13]
answers = [True, False, True, False]
for value, answer in zip(x, answers):
    if solution(value)== answer:
        print('정답')
    else:
        print('오답')
```
