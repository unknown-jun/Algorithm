```python
def solution(nums):
    nums_added=[]
    for i in range(len(nums)-2):
        for j in range(i+1,len(nums)-1):
            for k in range(j+1,len(nums)):
                nums_added.append(nums[i]+nums[j]+nums[k])

    cnt=0
    for i in nums_added:
        n = 1
        while True:
            n+=1
            if i==n :
                cnt+=1
                break
            elif i % n == 0 : 
                break
    return cnt
```

```python
def sieve(n):
    """
     
    """

    s = [0, 0] + [1] * (n - 1)

    for i in range(2, int(n**.5)+1):
        if s[i]:
            s[i*2::i] = [0] * len(s[i*2::i])
    return [i for i, v in enumerate(s) if v]

def solution(nums):
    primes = sieve(3000)

    count = 0
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            for k in range(j+1, len(nums)):
                if (nums[i] + nums[j] + nums[k]) in primes:
                    count += 1

    return count
```
