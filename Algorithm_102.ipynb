```python
def solution(nums):
    nums_added=[]
    for i in range(len(nums)-2):
        for j in range(i+1,len(nums)-1):
            for k in range(j+1,len(nums)):
                nums_added.append(nums[i]+nums[j]+nums[k])

    cnt=0
    for i in nums_added:
        n = 1
        while True:
            n+=1
            if i==n :
                cnt+=1
                break
            elif i % n == 0 : 
                break
    return cnt
```

```python
"""
itertools.combinations은
하나의 집합안에서 나올 수 있는 경우의 수를 추출할 수 있도록 도와주는 라이브러리다
itertools.combinations(nums, 3)은
nums 집합안의 원소를 3개씩 추출하게 한다.

for문에서는 소수를 추출하는 방식을 소개하고 있다.
num은 sum함수를 통해 combinaion 함수에서 뽑아낸 3개의 원소의 합을 객체화시킨다.
다음 for 문에서 range의 시작점을 2로 정함으로써 1로 나누었을때 0이 되는 경우를 제외시키고
끝 점을 num으로 정해 num-1까지 나눠지는 모든 경우를 제외시킨다.
그리고 이 중 나누어 떨어지는 경우가 있을 때에 break를 사용하여 for문을 빠져나가 다음 for문으로 진행시키고
모두가 break 되지 않고(모든 경우의 나머지가 1 이상일 경우) cnt를 하나 올린다.
"""
import itertools

def solution(nums):

    cnt = 0
    can = itertools.combinations(nums, 3)
    
    for i in can:
        num = sum(i)
        for j in range(2, num):
            if num % j == 0:
                break
        else:
            cnt += 1
    return cnt
```

```python
def sieve(n):
    """
    에라토스테네스의 체
    """

    s = [0, 0] + [1] * (n - 1)

    for i in range(2, int(n**.5)+1):
        if s[i]:
            s[i*2::i] = [0] * len(s[i*2::i])
    return [i for i, v in enumerate(s) if v]

def solution(nums):
    primes = sieve(3000)

    count = 0
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            for k in range(j+1, len(nums)):
                if (nums[i] + nums[j] + nums[k]) in primes:
                    count += 1

    return count
```
