```python
a ,b = [3, 3, 5], [5, 3, 3]
solutions =[12, 3, 12]

def solution(a, b):
    answer = 0
    tmp = []
    tmp.append(a)
    tmp.append(b)
    for i in range(sorted(tmp)[0], sorted(tmp)[1]+1):
        answer += i
    
    return answer

cnt=0
for i, z in zip(a, b):
    if solution(i, z) == solutions[cnt]:
        print('정답')
    else:
        print('오답')
    cnt += 1
```
