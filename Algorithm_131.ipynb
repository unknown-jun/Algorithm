```python
def solution(s, n):
    answer=''
    lower=list('abcdefghijklmnopqrstuvwxyz')
    upper=list('ABCDEFGHIJKLMNOPQRSTUVWXYZ')

    for i in s:
        if i in lower:
            answer += lower[(lower.index(i)+n) % 26]
        if i in upper:
            answer += upper[(upper.index(i)+n) % 26]
        if i ==' ':
            answer += ' '
            
    return answer
```

```python
s = ["AB", "z", "a B z"]
n = [1,1,4]
result = ["BC", "a", "e F d"]

for i in range(3):
    if solution(s[i],n[i]) == result[i]:
        print("")
```
