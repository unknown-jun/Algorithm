```python
def solution(n):
    n = list(str(n))
    answer = [True] * len(n)
    for idx, revidx in enumerate(range(len(n)-1, -1, -1)):
        answer[idx] = n[revidx]
    
    return list(map(int, answer))
```

```python
n = [12345,13247]

for i in n:
    print(solution(i))
```
