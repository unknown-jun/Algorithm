# 한줄로 풀수 있는 함수를 만들고자 함
1. str로 형 변환을 하여 list변환이 될 수 있게 함
2. 형변환을 한 list를 sorted를 하여 내림차순 정렬을 함
3. 이를 다시 int형으로 바꾸기 위하여 map 함수를 정의하여 int로 바꾸고 
4. 그 int를 list로 추가함

```python
def solution(n):
    return list(map(int, sorted(str(n), reverse=True)))
```

```python
n=12345
solution(n)
```

# 문제를 잘못 이해함

```python
n=13247
solution(n)
```

```python
n = list(str(n))
n
```

```python
return_value = [True] * len(n)
for idx, revidx in enumerate(range(len(n)-1,-1,-1)):
    return_value[idx] = n[revidx]
return_value
```

```python
return_value = list(map(int, return_value))
return_value
```

# 함수화

```python
def solution(n):
    n = list(str(n))
    answer = [True] * len(n)
    for idx, revidx in enumerate(range(len(n)-1, -1, -1)):
        answer[idx] = n[revidx]
    
    return list(map(int, answer))
```

```python
n = [12345,13247]

for i in n:
    print(solution(i))
```

# 다른 사람 풀이

```python
def solution(n):
    return list(map(int, reversed(str(n))))
# reversed라는 함수가 있는지 알지 못했다.
# 더더욱 기초에 충실해야 할듯!
```

```python
n = [12345,13247]

for i in n:
    print(solution(i))
```
