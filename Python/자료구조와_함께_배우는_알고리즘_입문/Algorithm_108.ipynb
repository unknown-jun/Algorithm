```python
def solution(numbers, hand):

    # 키패드를 좌표로 변경
    dic = {1:[0, 0], 2:[0, 1], 3:[0, 2],
           4:[1, 0], 5:[1, 1], 6:[1, 2],
           7:[2, 0], 8:[2, 1], 9:[2, 2],
           "*":[3, 0], 0:[3, 1], "#":[3, 2]}
    
    answer = ''
    left_s = dic['*']
    right_s = dic['#']
    
    for number in numbers:
        
        # 입력값으로 들어온 number를 좌표평면 상의 좌표로 변환
        now = dic[number]

        # 무조건 L인 경우를 조건으로 만들고 left의 위치를 갱신한다.
        if number in [1, 4, 7]:
            answer += 'L'
            left_s = now

        # 무조건 R인 경우를 조건으로 만들고 right의 위치를 갱신한다.
        elif number in [3, 6, 9]:
            answer += 'R'
            right_s = now

        # 2, 5, 8, 0들의 경우엔 맨하튼 거리를 계산한다.
        else:
            L_distance = abs(now[0]-left_s[0]) + abs(now[1]-left_s[1])
            R_distance = abs(now[0]-right_s[0]) + abs(now[1]-right_s[1])
            
            # R의 거리가 큰 경우 L을 반환하고 left의 위치를 갱신한다.
            if L_distance < R_distance:
                answer += 'L'
                left_s = now
                
            # L의 거리가 큰 경우 R을 반환하고 right의 위치를 갱신한다.
            elif R_distance < L_distance:
                answer += 'R'
                right_s = now
            
            # R의 거리와 L의 거리가 같은 경우 hand의 변수를 참조하여 결과값을 추가하고 위치를 갱신한다.
            else:
                if hand == "left":
                    answer += 'L'
                    left_s = now

                else:
                    answer +='R'
                    right_s = now
    return answer
```

```python
numbers_list =[[1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5], [7, 0, 8, 2, 8, 3, 1, 5, 7, 6, 2], [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]]
hands = ['right','left','right'] 
results = ["LRLLLRLLRRL", "LRLLRRLLLRR", "LLRLLRLLRL"]

for numbers, hand, result in zip(numbers_list, hands, results):
    
    if solution(numbers, hand) == result:
        print('정답입니다.')
    else:
        print(solution(numbers, hand), result)
        print('오답입니다.')
```
