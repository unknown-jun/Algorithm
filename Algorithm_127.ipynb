# Numeric 변수를 통한 풀이

```python
seoul = ["Jane", "Kim"]

def solution(seoul):
    answer = 0
    
    if 'Kim' in seoul:
        answer = seoul.index('Kim')
        
    return f"김서방은 {answer}에 있다"

solution(seoul)
```

# 굳이 Numeric 변수를 쓸 필요가 있을까?

```python
def solution(seoul):
    
    for i in seoul:
        if i == 'Kim':
            return f'김서방은 {seoul.index(i)}에 있다'

solution(seoul)
```

# 조금 더 줄일수 있지 않을까?

```python
def solution(seoul):
    
    return '김서방은 {}에 있다'.format(seoul.index('Kim'))

solution(seoul)
```
