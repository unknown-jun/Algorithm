# Numeric   

```python
seoul = ["Jane", "Kim"]

def solution(seoul):
    answer = 0
    
    if 'Kim' in seoul:
        answer = seoul.index('Kim')
        
    return f" {answer} "

solution(seoul)
```

#  Numeric    ?

```python
def solution(seoul):
    
    for i in seoul:
        if i == 'Kim':
            return f' {seoul.index(i)} '

solution(seoul)
```
