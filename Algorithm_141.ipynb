1. 입력된 수가 짝수라면 2로 나눈다.

```python
n = 4
while n != 1:
    if n % 2 == 0:
        n = n/2
        print(n)
```

2. 입력된 수가 홀수라면 3을 곱하고 1을 더함

```python
n = 626331
cnt = 0
while n != 1:
    
    if n % 2 == 0:
        n = n/2
    elif n % 2 !=0:
        n = n *3 +1
    cnt += 1

print(cnt)
```

함수화

```python
def solution(num):
    answer = 0
    while num != 1:

        if num % 2 == 0:
            num = num/2
        elif num % 2 !=0:
            num = num *3 +1
        answer += 1
    if answer < 500:
        return answer
    else:
        return -1
```

loop문이 500번이 넘을 경우 굳이 while문을 그 이상 돌릴 필요 없이  
종료 조건으로 실행시킴

```python
def solution(num):
    answer = 0
    while num != 1:

        if num % 2 == 0:
            num = num/2
        elif num % 2 !=0:
            num = num *3 +1
        answer += 1
        
        if answer == 500:
            return -1
            break
            
    return answer
```

```python
num = [6, 16, 626331]
result = [8, 4, -1]

for i, z in zip(num, result):
    if solution(i)==z:
        print('정답')
```
