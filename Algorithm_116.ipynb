2. 가운데 글자의 패턴은 무엇이 있을까?  
앞과 뒤의 글자의 수가 같다면 그 곳이 가운데 글자(ex. r /1  e  d /1 )

```python
string = 'red'
print(string[1]) # 0/ 1
print(string[-2]) # -1/ -2
```

```python
string = 'abcde'
print(string[2])  # 0/1
print(string[-3]) # -1/-2
```

```python
string = 'qwer'
print(string[1:3])   # 1 / 23 / 4
print(string[-3:-1]) # -4 / -3-2 / -1
```

4. s(입력값)의 길이를 구하고 그에 대해서 홀수이면 하나의 가운데 수를 반환하고,  
짝수이면 두개의 수를 반환해야 한다.

```python
s = "power overwheliming"

def solution(s):
    
    s_cnt = len(s)
    
    str_med = [ s.index(x) for x in s ]
    str_emd = int(np.median(str_med))
    
    if s_cnt %2 ==1:
        return s[str_emd]
    else:
        return s[str_emd:str_emd+2]
solution(s)
```

```python
def solution(s):
    
    s_cnt = len(s)
       
    if s_cnt %2 ==1:
        return s[int(s_cnt/2)]
    else:
        return s[int(s_cnt/2)-1:int(s_cnt/2)+1]
```
