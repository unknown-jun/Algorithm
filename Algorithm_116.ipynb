```python
s = "power overwheliming"

def solution(s):
    
    s_cnt = len(s)
    
    str_med = [ s.index(x) for x in s ]
    str_emd = int(np.median(str_med))
    
    if s_cnt %2 ==1:
        return s[str_emd]
    else:
        return s[str_emd:str_emd+2]
solution(s)
```
