```python
def solution(board, moves):
    
    answer = 0
    box = []

    for move in moves:
        for index in range(len(board)):
            if board[index][move-1] == 0:
                pass
            elif board[index][move-1] != 0:
                box.append(board[index][move-1])
                board[index][move-1] = 0
                break
            else:
                break

    for a in range(len(box)):
        for i in range(len(box)):
            try:
                if box[i] == box[i+1]:
                    answer += 1
                    del box[i+1]
                    del box[i]
            except:
                pass
    
    return answer * 2
```

```python
def solution(board, moves):
    basket = []
    answer = 0

    for i in moves:
        for j in range(len(board)):
            if board[j][i-1] != 0:
                basket.append(board[j][i-1])
                board[j][i-1] = 0

                if len(basket) > 1:
                    if basket[-1] == basket[-2]:
                        basket.pop(-1)
                        basket.pop(-1)
                        answer += 2     
                break

    return answer
```
