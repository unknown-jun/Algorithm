```python
arr = [[1, 2, 3, 4], [5, 5]]
return_value=[2.5, 5]

def solution(arr):
    answer = sum(arr) / len(arr)
    return answer

for i, z in zip(arr, return_value):
    if solution(i)==z:
        print("정답")
    else:
        print("오답")
```

```python
arr = [[1, 2, 3, 4], [5, 5]]
return_value=[2.5, 5]

def solution(arr):
    import numpy as np
    return np.mean(arr)

for i, z in zip(arr, return_value):
    if solution(i)==z:
        print("정답")
    else:
        print("오답")
```
