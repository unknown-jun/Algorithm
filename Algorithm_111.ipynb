```python
def solution(d, budget):
    
    import itertools
    temp = []
    
    for i in range(len(d)+1):
        
        d = sorted(d)
        d_itertool = itertools.combinations(d, i)
        for z in d_itertool:
            if sum(z) > budget:
                continue
            temp.append(i)
    
    return temp[-1]
```

```python
def solution(d, budget):
    temp = len(d)
    tmp_2 = []

    d = sorted(d)

    while True:
        if sum(d[:temp]) > budget:
            temp -= 1
        else: 
            tmp_2.append(temp)
            break
            
    return tmp_2[0]
```

```python
departs = [[1,3,2,5,4], [2,2,3,3]]
budgets =[9, 10]
results = [3, 4]

for depart, budget,result in zip(departs, budgets, results):
    
    if solution(depart, budget) == result:
        print('')
    else:
        print('')
```
