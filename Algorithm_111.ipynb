# 문제의 budget의 금액을 넘지 않는다 라는 조건 충족 시키기

```python
d = [1, 3, 2, 5, 4]
budget = 9
temp =[]

for i in range(len(d)+1):
    
    d_itertool = itertools.combinations(d, i)
    for z in d_itertool:
        if sum(z) > budget:
            continue
        temp.append(sum(z))
        
print(temp)
```

# 조건하의 최대값일 때의 itertools.combinations(d, i)에서 i 값을 찾아내기

```python
d = [1, 3, 2, 5, 4]
budget = 9
temp =[]

for i in range(len(d)+1):
    
    d = sorted(d)
    d_itertool = itertools.combinations(d, i)
    for z in d_itertool:
        if sum(z) > budget:
            continue
        temp.append(i)
print(temp[-1])
```

# 위의 코드를 함수화

```python
def solution(d, budget):
    
    import itertools
    temp = []
    
    for i in range(len(d)+1):
        
        d = sorted(d)
        d_itertool = itertools.combinations(d, i)
        for z in d_itertool:
            if sum(z) > budget:
                continue
            temp.append(i)
    
    return temp[-1]
```

# 프로그래머스 테스트
- 몇몇 테스트에서는 통과를 했지만 대부분 테스트에서 시간 초과 오류가 발생함  
- 이를 d가 많아질수록 iteration 수가 기하급수적으로 늘어나기 때문이라고 생각함  
- loop에 조건을 걸어 경우의 수 줄이기를 시도

# d를 정렬하여 모두를 더한 경우 budget보다 큰 경우를 iteration에서 제외시키기

```python
d = [1,3,2,5,4]
budget = 9

temp = len(d)
tmp_2 = []

d = sorted(d)

while True:
    if sum(d[:temp]) > budget:
        temp -= 1
    else: 
        tmp_2.append(temp)
        break
tmp_2
```

풀다보니 갑자기 맞춰버렸다.  
이 코드가 잘 작동하는 이유는 원소가 정렬이 되어 있을 경우, 만약 원소들의 합이 예산보다 크면 그 경우 어떠한 조합을 하던지 간에 예산보다 크다  
그렇기 때문에 원소들의 합이 예산보다 작거나 같은 경우가 가장 최적의 방법이라고 생각했다.

```python
def solution(d, budget):
    temp = len(d)
    tmp_2 = []

    d = sorted(d)

    while True:
        if sum(d[:temp]) > budget:
            temp -= 1
        else: 
            tmp_2.append(temp)
            break
            
    return tmp_2[0]
```

```python
departs = [[1,3,2,5,4], [2,2,3,3]]
budgets =[9, 10]
results = [3, 4]

for depart, budget,result in zip(departs, budgets, results):
    
    if solution(depart, budget) == result:
        print('정답')
    else:
        print('오답')
```
