# 정수의 제곱근이라면 float형으로 형변환이 되어져 나올때 끝자리가 0이라는 점에 착안
- str로 형변환하여 맨 마지막을 슬라이싱
- 슬라이싱을 조건으로 지정하여 '0'이면 규칙이 되게 만듦

```python
import numpy as np

a = np.sqrt(121)
str(a)[-1]
```

```python
n=[121,3]

for i in n:
    if str(np.sqrt(i))[-1]=='0':
        print('2')
    else:
        print('1')
```

# 문제에 맞는 조건 생성

```python
n=[121,3]

for i in n:
    if str(np.sqrt(i))[-1]=='0':
        print(int(np.sqrt(i)+1)**2)
    else:
        print(-1)
```

```python
def solution(n):
    import numpy as np
    
    if str(np.sqrt(n))[-1]=='0':
        return int(np.sqrt(n)+1)**2
    else:
        return -1
```

```python
n=[121,3]
for i, z in zip(n, [144, -1]):
    if solution(i) == z:
        print('정답')
```
