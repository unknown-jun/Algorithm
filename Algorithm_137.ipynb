```python
def solution(n):
    import numpy as np
    
    if str(np.sqrt(n))[-1]=='0':
        return int(np.sqrt(n)+1)**2
    else:
        return -1
```

```python
n=[121,3]
for i, z in zip(n, [144, -1]):
    if solution(i) == z:
        print('')
```
