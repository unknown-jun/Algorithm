# 진법변환 코드 작성

```python
# 나누기 연산 후 소수점 이하의 수를 버리고 정수 부분의 수만 구하는 코드
n = 45

a_1 = n // 3
a_2 = a_1 // 3
a_3 = a_2 // 3
a_4 = a_3 // 3

print(a_1)
print(a_2)
print(a_3)
print(a_4)
```

```python
# 나누기 연산 중 몫이 0일 때 멈추는 루프문 작성
n = 45

while n != 0:
    print(n)
    n = n // 3
```

```python
# 나누기 연산 중 나머지를 추가하고 몫이 0일 때 멈추는 루프문 작성 <반전 3진법>

n = 125
remainder = ''

while n != 0:
    remainder += str(n % 3)
    n = n // 3
    
remainder
```

# 10진법으로 바꾸기

```python
Decimal = 0
cnt = len(remainder)-1

for i in remainder:
    Decimal += int(i) * 3**cnt
    cnt -= 1
    
Decimal
```

# 함수화 시키기

```python
def solution(n):
    remainder = ''

    while n != 0:
        remainder += str(n % 3)
        n = n // 3
    
    Decimal = 0
    cnt = len(remainder)-1

    for i in remainder:
        Decimal += int(i) * 3**cnt
        cnt -= 1
    
    return Decimal
```
