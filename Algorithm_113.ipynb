```python
def solution(n):
    remainder = ''

    while n != 0:
        remainder += str(n % 3)
        n = n // 3
    
    Decimal = 0
    cnt = len(remainder)-1

    for i in remainder:
        Decimal += int(i) * 3**cnt
        cnt -= 1
    
    return Decimal
```
