```python
def solution(new_id):
    import re
    
    st = new_id.lower()
    st = re.sub(r'[^a-z0-9\-_\.]', '', st)
    st = re.sub(r'\.+','.', st)
    st = re.sub(r'\.$', '', st)
    st = re.sub(r'^\.', '', st)
    st = 'a' if len(st) == 0 else st[:15]
    st = re.sub(r'\.$', '', st)

    while len(st) < 3:
        st = st + st[-1]

    return st
```

```python
def solution(new_id):
    
    answer = ''
    # 1단계 new_id의 모든 대문자를 대응되는 소문자로 치환
    new_id = new_id.lower( )
    
    # 2단계 new_id에서 알파벳 소문자, 숫자, 빼기(-), 밑줄(_), 마침표(.)를 제외한 모든 문자를 제거함
    for c in new_id:
        if c.isalpha() or c.isdigit() or c in ['-','_','.']:
            answer += c  # 빈 문자열에 더하기(+)를 하면 해당 문자열이 추가가 된다.
    
    # 3단계 new_id에서 마침표(.)가 2번 이상 연속된 부분을 하나의 마침표(.)로 치환한다.
    while '..' in answer:
        answer = answer.replace('..','.')
        
    # 4단계 new_id에서 마침표(.)가 처음이나 끝에 위치한다면 제거함
    print(answer)
    if answer[0] == '.':
        answer = answer[1:] if len(answer) > 1 else '.'
                            # 이 if절이 없으면 "=.="와 같은 경우 다음 if answer[-1] == '.'절에서 만족하는 경우가 없어지기 때문에 오류가 발생
    if answer[-1] == '.':
        answer = answer[:-1]
    
    # 5단계 new_id가 빈 문자열이라면, new_id에 "a"를 대입한다,
    if answer == '':
        answer = 'a'
        
    # 6단계 new_id의 길이가 16자 이상이면, new_id의 첫 15개의 문자를 제외한 나머지 문자들을 모두 제거한다.
    #      만약 제거 후 마침표(.)가 new_id의 끝에 위치한다면 끝에 위치한 마침표(.)문자를 제거함
    if len(answer)>15:
        answer = answer[:15]
        if answer[-1] == '.':
            answer = answer[:-1]
    
    # 7단계 new_id의 길이가 2자 이하라면, new_id의 마지막 문자를 new_id의 길이가 3이 될 때까지 반복해서 끝에 뭍임
    while len(answer) < 3:
        answer += answer[-1]

    return answer
```
