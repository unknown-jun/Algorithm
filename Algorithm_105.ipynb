```python
def solution(new_id):
    import re
    
    st = new_id.lower()
    st = re.sub(r'[^a-z0-9\-_\.]', '', st)
    st = re.sub(r'\.+','.', st)
    st = re.sub(r'\.$', '', st)
    st = re.sub(r'^\.', '', st)
    st = 'a' if len(st) == 0 else st[:15]
    st = re.sub(r'\.$', '', st)

    while len(st) < 3:
        st = st + st[-1]

    return st
```
